```python
# ===============================
# DAY 15 — FINAL CONFIG LOCK
# 6-Channel Symmetric Configuration (Final Choice)
# ===============================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt

# -------------------------------
# Load Day 11 Channel Results
# -------------------------------
file_path = "/content/drive/MyDrive/EEG/day11_channel_reduction_results.csv"
df = pd.read_csv(file_path)

# -------------------------------
# FINAL LOCKED PARAMETERS
# -------------------------------
final_channels = 6
selected_noise = [0, 1]   # clean and noisy condition (σ = 1)

# Freeze everything — no more tuning
FINAL_CSP_COMPONENTS = 4
FINAL_NOISE_STD = 1.0

# -------------------------------
# Filter Final Configuration
# -------------------------------
df_final = df[
    (df['channels'] == final_channels) &
    (df['noise_std'].isin(selected_noise))
].copy()

if df_final.empty:
    raise ValueError("No 6-channel results found. Ensure Day 11 CSV includes 6-channel symmetric configuration.")

# -------------------------------
# Compute Final Delta
# -------------------------------
df_final['delta_acc'] = df_final['csp_mean_acc'] - df_final['base_mean_acc']

# -------------------------------
# Save Final Locked Results
# -------------------------------
output_csv = "/content/drive/MyDrive/EEG/day15_final_6ch_results.csv"
df_final.to_csv(output_csv, index=False)

print("FINAL LOCKED CONFIGURATION (6 Channels)")
print(df_final)

# ===============================
# Plot 1 — Final Comparison Plot
# ===============================
plt.figure()

plt.errorbar(df_final['noise_std'],
             df_final['csp_mean_acc'],
             yerr=df_final['csp_std'],
             marker='o',
             label='CSP (6ch)')

plt.errorbar(df_final['noise_std'],
             df_final['base_mean_acc'],
             yerr=df_final['base_std'],
             marker='s',
             label='Baseline (6ch)')

plt.xlabel("Noise STD")
plt.ylabel("Accuracy")
plt.title("Final Locked Evaluation: 6-Channel CSP vs Baseline")
plt.legend()
plt.grid(True)

plot_path = "/content/drive/MyDrive/EEG/day15_final_comparison_plot.png"
plt.savefig(plot_path)
plt.show()

# ===============================
# Plot 2 — Final Delta Plot
# ===============================
plt.figure()

plt.bar(df_final['noise_std'], df_final['delta_acc'])

plt.xlabel("Noise STD")
plt.ylabel("CSP - Baseline Accuracy")
plt.title("Final Delta Accuracy (6 Channels)")
plt.grid(True)

delta_plot_path = "/content/drive/MyDrive/EEG/day15_final_delta_plot.png"
plt.savefig(delta_plot_path)
plt.show()

print("Day 15 FINAL configuration locked.")
print("Saved files:")
print(output_csv)
print(plot_path)
print(delta_plot_path)
```
