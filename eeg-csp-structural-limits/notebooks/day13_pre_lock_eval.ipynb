# Day 13 — 5-Channel Reduced Configuration  
## Pre-Lock Structural Evaluation

### Objective
Evaluate an aggressively reduced 5-channel asymmetric configuration prior to final symmetric locking.

Channels:
FC5, FC6, Cz, C3, C4

Noise levels:
- STD = 0 (clean)
- STD = 1 (noisy)

CSP components: 4

---

## Results

### Clean Condition (STD = 0)
- CSP: 0.4921
- Baseline: 0.5004
- Δ = -0.0083

### Noisy Condition (STD = 1)
- CSP: 0.4971
- Baseline: 0.5004
- Δ = -0.0033

---

## Observations
- CSP underperforms baseline in both conditions.
- Negative delta indicates loss of discriminative advantage.
- Noise injection slightly improves CSP mean accuracy but does not surpass baseline.
- Baseline accuracy remains stable across noise levels.
- Performance differences remain small (<1% absolute).

---

## Structural Interpretation
With only 5 channels, CSP loses its discriminatory advantage.

The asymmetric reduction likely disrupts covariance structure required for stable spatial filtering.

Noise at STD = 1 reduces the performance gap slightly, but CSP remains inferior.

This configuration lies outside the dominance regime identified at 8 channels.

```python
# ===============================
# DAY 13 — 5-Channel Pre-Lock Evaluation
# Configuration: FC5, FC6, Cz, C3, C4
# Purpose: Evaluate reduced asymmetric configuration prior to final symmetric lock
# ===============================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt

# -------------------------------
# Load Day 11 Channel Results
# -------------------------------
file_path = "/content/drive/MyDrive/EEG/day11_channel_reduction_results.csv"
df = pd.read_csv(file_path)

# -------------------------------
# Configuration Parameters
# -------------------------------
eval_channels = 5
selected_noise = [0, 1]   # clean and noisy condition (σ = 1)

CSP_COMPONENTS = 4
NOISE_STD = 1.0

# -------------------------------
# Filter 5-Channel Configuration
# -------------------------------
df_eval = df[
    (df['channels'] == eval_channels) &
    (df['noise_std'].isin(selected_noise))
].copy()

if df_eval.empty:
    raise ValueError("No 5-channel results found. Ensure Day 11 CSV includes the 5-channel configuration.")

# -------------------------------
# Compute Delta
# -------------------------------
df_eval['delta_acc'] = df_eval['csp_mean_acc'] - df_eval['base_mean_acc']

# -------------------------------
# Save Evaluation Results
# -------------------------------
output_csv = "/content/drive/MyDrive/EEG/day13_5ch_eval_results.csv"
df_eval.to_csv(output_csv, index=False)

print("DAY 13 — 5-Channel Pre-Lock Evaluation")
print(df_eval)

# ===============================
# Plot 1 — Comparison Plot
# ===============================
plt.figure()

plt.errorbar(df_eval['noise_std'],
             df_eval['csp_mean_acc'],
             yerr=df_eval['csp_std'],
             marker='o',
             label='CSP (5ch)')

plt.errorbar(df_eval['noise_std'],
             df_eval['base_mean_acc'],
             yerr=df_eval['base_std'],
             marker='s',
             label='Baseline (5ch)')

plt.xlabel("Noise STD")
plt.ylabel("Accuracy")
plt.title("5-Channel Evaluation: CSP vs Baseline")
plt.legend()
plt.grid(True)

plot_path = "/content/drive/MyDrive/EEG/day13_5ch_comparison_plot.png"
plt.savefig(plot_path)
plt.show()

# ===============================
# Plot 2 — Delta Plot
# ===============================
plt.figure()

plt.bar(df_eval['noise_std'], df_eval['delta_acc'])

plt.xlabel("Noise STD")
plt.ylabel("CSP - Baseline Accuracy")
plt.title("5-Channel Delta Accuracy")
plt.grid(True)

delta_plot_path = "/content/drive/MyDrive/EEG/day13_5ch_delta_plot.png"
plt.savefig(delta_plot_path)
plt.show()

print("Day 13 evaluation complete.")
print("Saved files:")
print(output_csv)
print(plot_path)
print(delta_plot_path)
```
